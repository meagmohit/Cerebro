```python
##Importing Libraries

import numpy as np
import scipy.io
import os
from sklearn.decomposition import FastICA
import mne
from mne.time_frequency import psd_multitaper
import matplotlib.pyplot as plt
from scipy import stats

mne.set_log_level('ERROR')
```

```python
## Parameters

chan_list = ['Fp1','F3','F7','C3','T7','P3','P7','O1','Pz','Fp2','Fz','F4','F8','Cz','C4','T8','P4','P8','O2']

selected_chan = [1,3,8,10,11,13,14]
total_chan = len(selected_chan)

total_sub = 14  # Anything from 1-14
total_prod = 10 
freq = 256.0
time_len = 768

time = [(x-freq)*1.0/freq for x in xrange(1,time_len+1)]
time = np.array(time)

n200_ind = [idx for idx, t_ind in enumerate(time) if (t_ind>=0.2 and t_ind<=0.3)]
n200_ind = np.array(n200_ind)

erp_ind = [idx for idx, t_ind in enumerate(time) if (t_ind>=0.2 and t_ind<=0.3)]
erp_ind = np.array(erp_ind)
erp_len = len(erp_ind)
```

```python
# Loading data
data_dict = {}
data_dir = 'data/'
list_of_files = [f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))]
for curr_file in list_of_files:
    if '_data.mat' in curr_file:
        sub_id = int(curr_file[1:3])
        data_dict[sub_id] = scipy.io.loadmat(os.path.join(data_dir,curr_file))
    if curr_file == 'WA2.mat':
        WA2 = scipy.io.loadmat(os.path.join(data_dir,curr_file))['WA2']
        WA2 = np.delete(WA2,11,0)  # to remove 12th subject
    if curr_file == 'outChoices.mat':
        outChoices = scipy.io.loadmat(os.path.join(data_dir,curr_file))['out']
# WA2 contains 14(sub) x 10 (prod) : value represents # of times the product was chosen total
# outChoices : column 0: if prod_i was chosen? column 7: sub_id*100+prod_1 , col 8: sub_id*100+prod_2
choices = np.zeros([total_sub+1, total_prod+1, total_prod+1])
for idx in range(outChoices.shape[0]):
    sub_id = int(outChoices[idx, 7]//100)
    sub_id_2 = int(outChoices[idx, 8]//100)
    if sub_id == 12:
        continue
    if sub_id > 12:
        sub_id = sub_id - 1
        sub_id_2 = sub_id_2 - 1
    assert sub_id>0 and sub_id <= (total_sub+1) and sub_id == sub_id_2, "Error 1: error decoding"+str(sub_id)
    prod_1 = int(outChoices[idx, 7]%100)
    prod_2 = int(outChoices[idx, 8]%100)
    assert prod_1 > 0 and prod_1 <= total_prod and  prod_2 > 0 and prod_2 <= total_prod, "Error 2: error decoding "+str(prod_2)
    if prod_1 > prod_2 or prod_1==prod_2:
        print "check it baby", prod_1, prod_2
    if outChoices[idx, 0] == 0:
        choices[sub_id, prod_1, prod_2] = choices[sub_id, prod_1, prod_2] + 1
    elif outChoices[idx, 0] == 1:
        choices[sub_id, prod_2, prod_1] = choices[sub_id, prod_2, prod_1] + 1
        
        
pref = np.zeros([total_sub+1, total_prod+1])
pref[1:,1:] = WA2
#data_dict[sub_id]['sig'], ['lab'] contains original eeg signals (25x768x500) and labels (1x500)
# What are channels?
```

```python
ncomp = 1
n200_val = np.zeros([total_sub+1, total_prod+1])
ersp_val = np.zeros([total_sub+1, total_prod+1])
feat_val = np.zeros([total_sub+1, total_prod+1])
for sub_id in range(1,total_sub+1):
    sig = data_dict[sub_id]['sig']
    lab = data_dict[sub_id]['lab'].flatten()
    for prod_id in range(1,total_prod+1):
        sig_prod =  sig[selected_chan, :, np.argwhere(lab==prod_id)]     
        avg_sig_chan = np.transpose(np.mean(sig_prod, axis=0))
        # compute ICA
        ica_sig = FastICA(n_components=ncomp)
        S_ = ica_sig.fit_transform(avg_sig_chan)  # Get the estimated sources
        A_sig = ica_sig.mixing_  # Get estimated mixing matrix
        A_sig_norm = np.linalg.norm(A_sig,axis=0)
        S_ = S_*A_sig_norm
    
        if sum(A_sig)<0:
            S_ = -1*S_
        
        n200val =  np.mean(S_[n200_ind])
        featval = min(S_[erp_ind])
        
        info = mne.create_info(ch_names=['ica'], sfreq=freq, ch_types=['eeg'])
        raw = mne.io.RawArray(np.transpose(S_[256:]), info)
        psds, freqs = psd_multitaper(raw, low_bias=True, tmin=0.1, tmax=0.5, fmin=13, fmax=26, proj=True, n_jobs=1)
        raw2 = mne.io.RawArray(np.transpose(S_), info)
        psds2, freqs2 = psd_multitaper(raw2, low_bias=True, tmin=0.5, tmax=1.0, fmin=13, fmax=26, proj=True, n_jobs=1)
        erspval = np.mean(10*np.log10(psds)) - np.mean(10*np.log10(psds2))

        n200_val[sub_id, prod_id] = -1*abs(n200val)
        ersp_val[sub_id, prod_id] = -1*erspval
        feat_val[sub_id, prod_id] = featval

        
```

```python
# Computing Correlation for n200/ersp/feat

data_n200 = n200_val[1:,1:].flatten()
data_ersp = ersp_val[1:,1:].flatten()
data_feat = feat_val[1:,1:].flatten()

print "N200 and ERSP correlation:", scipy.stats.pearsonr(data_n200, data_ersp)

import seaborn as sns; sns.set()
np.corrcoef(data_ersp, data_feat)

sns.set(font_scale=1.4)
map_data = np.array([[1.0, 0.7349, 0.0025],[0.7349, 1.0, -0.229],[0.0025, -0.229, 1.0]])
labels = ['N200 \n mean', 'N200 \n minima', 'ERSP']
ax = sns.heatmap(map_data, vmin=-1, vmax=1.0, center=0, annot=True, linewidths=.5, cmap='PuOr', xticklabels=labels, yticklabels=labels)
# ax.tick_params(direction='out', length=6, width=2, colors='r', grid_color='r', grid_alpha=0.5)
```

```python
# Pairwise Evaluation

total = 0
# score_n200, score_feat, score_ersp, score_decn = 0, 0, 0, 0
dist_pref = 47
score_metric = np.zeros([4,total_sub+1])

for sub_id in range(1,total_sub+1):
    total = 0
    score_n200, score_feat, score_ersp, score_decn = 0, 0, 0, 0
    for prod_i in range(1,total_prod+1):
        for prod_j in range(prod_i+1,total_prod+1):
            if abs(pref[sub_id, prod_i] - pref[sub_id, prod_j]) > dist_pref:
                
                total = total + 1
                ground_truth = (pref[sub_id, prod_i] > pref[sub_id, prod_j]) # True means "prod_j" is preferred
                
                count = 0
                pred_n200 = n200_val[sub_id, prod_i] > n200_val[sub_id, prod_j]
                if (pred_n200 == ground_truth):
                    score_n200 = score_n200 + 1 
                    count = count + 1
                    
                pred_feat = feat_val[sub_id, prod_i] > feat_val[sub_id, prod_j]
                if (pred_feat == ground_truth):
                    score_feat = score_feat + 1 
                    count = count + 1
                    
                pred_ersp = ersp_val[sub_id, prod_i] > ersp_val[sub_id, prod_j]
                if (pred_ersp == ground_truth):
                    score_ersp = score_ersp + 1 
                    count = count + 1
                
                if count > 1:
                    score_decn = score_decn + 1
                    
    score_metric[0, sub_id] = score_n200*1.0/total
    score_metric[1, sub_id] = score_feat*1.0/total
    score_metric[2, sub_id] = score_ersp*1.0/total
    score_metric[3, sub_id] = score_decn*1.0/total
    
acc_mean = np.mean(score_metric[:,1:], axis=1)
acc_std = np.std(score_metric[:,1:], axis=1)
# print "N200 Accuracy: ", "Feat Accuracy: ", "ERSP Accuracy: ", "Decision Accuracy: "
# print "Mean: ", np.mean(score_metric[:,1:], axis=1)
# print "Std: ", np.std(score_metric[:,1:], axis=1)

method = ['N200 mean', 'N200 minima', 'ERSP', 'Decision']

print "\033[1m Fixed Rule based Pairwise Evaluation \033[0m"
print "Mean Accuracy:"
for idx,item in enumerate(acc_mean):
    print method[idx], item
print "Standard Deviation:"    
for idx,item in enumerate(acc_std):
    print method[idx], item
```

```python
## Ranking 3 prooducts

rank_metrics = np.zeros([5,3]) # rows: N200, feat, ersp, decn1, 4 random col: avg dist, tau, NDCG
total = 0

def tau_score(t_rank, p_rank):
    tau, _ = stats.kendalltau(p_rank, t_rank)
    return tau

# Function to calculate NDCG score
def ndcg_score(rel, t_rank, p_rank):
    rel = np.array(rel)
    dcg = sum((rel)*1.0/np.log2(p_rank + 1))
    idcg = sum((rel)*1.0/np.log2(t_rank + 1))
    return dcg*1.0/idcg


for sub_id in range(3,4):#1,total_sub+1):
    for prod_i in range(1,total_prod+1):
        for prod_j in range(prod_i+1,total_prod+1):
            for prod_k in range(prod_j+1,total_prod+1):
#                 if not ((pref[sub_id, prod_i] == pref[sub_id, prod_j]) or (pref[sub_id, prod_i] == pref[sub_id, prod_k]) or (pref[sub_id, prod_j] == pref[sub_id, prod_k])):
                    total = total + 1
                    ground_pref = [pref[sub_id, prod_i], pref[sub_id, prod_j], pref[sub_id, prod_k]]   # [rank_i, rank_j, rank_k]
                    ground_ordr = np.argsort(ground_pref)
                    ground_rank = 3 - np.argsort(ground_ordr)  # rank 1 means top, rank 3 means bottom
                    true_rel = ground_pref
                    
                    
                    pred_n200 = [n200_val[sub_id, prod_i], n200_val[sub_id, prod_j], n200_val[sub_id, prod_k]]
                    ordr_n200 = np.argsort(pred_n200)
                    rank_n200 = 3 - np.argsort(ordr_n200)

                    pred_feat = [feat_val[sub_id, prod_i], feat_val[sub_id, prod_j], feat_val[sub_id, prod_k]]
                    ordr_feat = np.argsort(pred_feat)
                    rank_feat = 3 - np.argsort(ordr_feat)

                    pred_ersp = [ersp_val[sub_id, prod_i], ersp_val[sub_id, prod_j], ersp_val[sub_id, prod_k]]
                    ordr_ersp = np.argsort(pred_ersp)
                    rank_ersp = 3 - np.argsort(ordr_ersp)
                
                    # Computing for decision classifier 1
                    
                    if np.mean(rank_n200==rank_ersp)==1:
                        rank_decn = rank_n200
                    elif np.mean(rank_feat==rank_ersp)==1:
                        rank_decn = rank_feat
                    elif np.mean(rank_n200==rank_feat)==1:
                        rank_decn = rank_n200
                    else:
                        count = [0, 0, 0]
                        count[rank_n200[0]-1] = count[rank_n200[0]-1] + 2
                        count[rank_n200[1]-1] = count[rank_n200[1]-1] + 1
                        count[rank_feat[0]-1] = count[rank_feat[0]-1] + 2
                        count[rank_feat[1]-1] = count[rank_feat[1]-1] + 1
                        count[rank_ersp[0]-1] = count[rank_ersp[0]-1] + 2
                        count[rank_ersp[1]-1] = count[rank_ersp[1]-1] + 1
                        ordr_decn = np.argsort(count)
                        rank_decn = 3 - np.argsort(ordr_decn)
                        
                    # Computing for Random
                    pred_rand = np.random.uniform(0,1,[1,3])
                    ordr_rand = np.argsort(pred_rand)
                    rank_rand = 3 - np.argsort(ordr_rand)
                    rank_rand = rank_rand.flatten()
                    
                    all_ranks = [rank_n200, rank_ersp, rank_feat, rank_decn, rank_rand]
                    for idx_r, pred_rank in enumerate(all_ranks):
                        rank_metrics[idx_r,0] = rank_metrics[idx_r,0] + np.mean(abs(ground_rank - pred_rank))
                        rank_metrics[idx_r,1] = rank_metrics[idx_r,1] + tau_score(ground_rank, pred_rank)
                        rank_metrics[idx_r,2] = rank_metrics[idx_r,2] + ndcg_score(true_rel, ground_rank, pred_rank)
                    
                    
                    
                    
                    
print "\033[1m Fixed Rule based Ranking \033[0m"
print "Metrics:        MHD,         Tau,        NDCG"
methods = ['N200 Mean  ', 'ERSP       ', 'N200 minima', 'Decision   ', 'Random     ']
for idx, item in enumerate(rank_metrics):
    print methods[idx], item/total
                    
# print rank_metrics/total
# print total
```
